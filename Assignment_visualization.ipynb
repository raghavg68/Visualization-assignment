```python
# Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

categories = ['A', 'B', 'C', 'D', 'E']
num_samples = 1000

data = pd.DataFrame({
    'category': np.random.choice(categories, size=num_samples),
    'value': np.random.normal(10, 2, num_samples)
})

average_values = data.groupby('category')['value'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(average_values['category'], average_values['value'], color='skyblue')
plt.title('Average Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Average Value')
plt.ylim(0, average_values['value'].max() + 2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(42)
n_samples = 100
n_features = 5
X = np.random.randn(n_samples, n_features)
X[:, 1] = X[:, 0] + np.random.randn(n_samples) * 0.5
X[:, 2] = X[:, 0] + np.random.randn(n_samples) * 0.8
X[:, 3] = X[:, 0] + np.random.randn(n_samples) * 0.3
corr_matrix = np.corrcoef(X.T)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap: Correlation Matrix')
plt.show()
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
# Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Scatter3d(
    x=df['X'],
    y=df['Y'],
    z=df['Z'],
    mode='markers',
    marker=dict(
        size=5,
        color=df['Z'],
        colorscale='Viridis',
        opacity=0.8
    )
)])

fig.update_layout(title='3D Scatter Plot')
fig.show()
```

```python
# . Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

fig = px.violin(df, x="Grade", y="Score", box=True, points="all")
fig.update_layout(title='Distribution of Scores Across Grades')
fig.show()
```

```python
# Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

pivot_df = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

fig = go.Figure(data=go.Heatmap(
    z=pivot_df.values,
    x=pivot_df.columns,
    y=pivot_df.index,
    colorscale='Viridis'
))

fig.update_layout(
    title='Sales Heatmap: Variation Across Months and Days',
    xaxis_title='Month',
    yaxis_title='Day'
)
fig.show()
```

```python
# Using the given x and y data, generate a 3D surface plot to visualize the function

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot: z = sin(sqrt(x^2 + y^2))',
                  scene = dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'))
fig.show()
```

```python
# Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

fig = px.scatter(df, x="GDP", y="Population", size="Population", color="Country",
                 hover_name="Country", size_max=100)

fig.update_layout(title='Bubble Chart: Country GDP vs Population',
                  xaxis_title='GDP',
                  yaxis_title='Population')
fig.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.layouts import column
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral10
from bokeh.io import output_notebook, show
```

```python
# Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p1 = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p1.line(x, y, line_width=2)
show(p1)
```

```python
# Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

N = 100
x = np.random.random(N)
y = np.random.random(N)
sizes = np.random.randint(10, 50, N)
colors = np.random.choice(Spectral10, N)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p2 = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')
p2.circle('x', 'y', size='sizes', color='colors', alpha=0.5, source=source)
show(p2)
```

```python
# Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p3 = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")
p3.vbar(x=fruits, top=counts, width=0.9)
p3.xgrid.grid_line_color = None
p3.y_range.start = 0
show(p3)
```

```python
# Create a Bokeh histogram to visualize the distribution of the given data

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p4 = figure(title="Histogram", x_axis_label='Value', y_axis_label='Count')
p4.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)
show(p4)
```

```python
# Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p5 = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1), toolbar_location=None, tools="")
p5.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, palette="Spectral11")
show(p5)
```
